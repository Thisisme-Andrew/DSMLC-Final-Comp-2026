## Wind Farm A

```python
df_a = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ";") \
    .load("/Volumes/workspace/wind-turbine-bronze/wind-farm-a/datasets")
display(df_a)
```

```python
df_a.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("workspace.`wind-turbine-silver`.`wind-farm-a`")
```
